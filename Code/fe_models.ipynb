# Starter Notebook: FE + CLR and Pooled OLS Models

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from linearmodels.panel import PanelOLS

def clr_transform(df, cols):
    X = df[cols].replace(0, np.nan)
    gm = np.exp(np.log(X).mean(axis=1))
    clr_df = np.log(X.div(gm, axis=0))
    clr_df.columns = ['clr_' + c for c in cols]
    return pd.concat([df, clr_df], axis=1)
```
